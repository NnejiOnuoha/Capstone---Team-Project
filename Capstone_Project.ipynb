```python
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon
import nltk
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Function to get sentiment polarity score
def get_sentiment(text):
    scores = sid.polarity_scores(text)
    return scores['compound']

# Apply sentiment analysis to the text responses
df_cleaned['Sentiment Score'] = df_cleaned['What do you think about searching for a roommate or accomodations through agents?'].apply(get_sentiment)

# Visualize sentiment distribution
plt.figure(figsize=(8, 5))
sns.histplot(data=df_cleaned, x='Sentiment Score', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Define a function to categorize sentiment scores
def categorize_sentiment(score):
    if score == 0:
        return 'Neutral'
    elif 0 < score <= 0.5:
        return 'Positive'
    elif 0.5 < score <= 1:
        return 'Very Positive'
    elif -0.5 <= score < 0:
        return 'Negative'
    elif -1 <= score < -0.5:
        return 'Very Negative'

# Apply the function to create a new column for sentiment categories
df_cleaned['Sentiment Category'] = df_cleaned['Sentiment Score'].apply(categorize_sentiment)

# Visualize sentiment distribution based on categories
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cleaned, x='Sentiment Category', palette='coolwarm')
plt.title('Distribution of Sentiment Categories')
plt.xlabel('Sentiment Category')
plt.ylabel('Frequency')
plt.show()
```

```python
priority_column = 'How will you prioritize the following factors when searching for an accommodation or roommate? Location, budget, lifestyle, preference, religion? Arrange them from highest to least. '

# Drop rows with missing values in the priority column
df_cleaned.dropna(subset=[priority_column], inplace=True)

# Split the responses into individual factors
df_cleaned['Priorities'] = df_cleaned[priority_column].str.split(', ')

# Flatten the list of priorities, handling NaN values
all_priorities = [factor for sublist in df_cleaned['Priorities'].dropna() for factor in sublist]

# Count the occurrences of each factor
priority_counts = pd.Series(all_priorities).value_counts()

# Define the order of factors
priority_order = ['Location', 'Budget', 'Lifestyle', 'Preference', 'Religion']

# Sort the factors based on their counts and priority order
sorted_priorities = [factor for factor in priority_order if factor in priority_counts.index]

# Plot the priority order
plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_priorities)), priority_counts[sorted_priorities], color='skyblue')
plt.yticks(range(len(sorted_priorities)), sorted_priorities)
plt.xlabel('Count')
plt.ylabel('Priority')
plt.title('Priority of Factors when Searching for Accommodation or Roommate')
plt.gca().invert_yaxis()  # Invert y-axis to display highest priority at the top
plt.show()
```

```python
app_usage_column = 'Have you ever used an app to search for an accommodation or roommate? If yes what app did you use and how was your experience using it? '

# Filter out responses other than "Yes" and "No"
app_usage_counts = df_cleaned[app_usage_column].value_counts().loc[['Yes', 'No']]

# Plot the doughnut chart with custom colors
plt.figure(figsize=(8, 6))
colors = ['skyblue', 'lightcoral']  # Using light red and sky blue colors
plt.pie(app_usage_counts, labels=app_usage_counts.index, autopct='%1.1f%%', colors=colors, wedgeprops=dict(width=0.3))
plt.title('Respondents Usage of Apps for Searching Accommodations or Roommates')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon
import nltk
nltk.download('vader_lexicon')

# Assuming the column name containing participants' sentiments is 'Interest in Using App'
interest_column = "If you haven’t used such an app or come across it, will you love to use an app that caters for your rooms and accommodation needs?"

# Instantiate the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define a function to categorize sentiment scores
def categorize_sentiment(score):
    if score == 0:
        return 'Neutral'
    elif 0 < score <= 0.5:
        return 'Positive'
    elif 0.5 < score <= 1:
        return 'Very Positive'
    elif -0.5 <= score < 0:
        return 'Negative'
    elif -1 <= score < -0.5:
        return 'Very Negative'

# Define a function to get sentiment polarity score
def get_sentiment(text):
    return sid.polarity_scores(text)['compound']

# Apply the function to the interest column to get sentiment scores
df_cleaned['Interest Sentiment Score'] = df_cleaned[interest_column].apply(get_sentiment)

# Apply the function to categorize sentiment scores
df_cleaned['Interest Sentiment Category'] = df_cleaned['Interest Sentiment Score'].apply(categorize_sentiment)

# Plot the sentiment distribution based on categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df_cleaned, x='Interest Sentiment Category', palette='coolwarm')
plt.title('Sentiment Distribution towards Using Accommodation App')
plt.xlabel('Sentiment Category')
plt.ylabel('Frequency')
plt.show()
```

```python
# Column containing the willingness to pay information
willingness_to_pay_column = 'Are you willing to pay a token  or subscribe to an app to take away that burden of search from you?'

# Extract the sentiment information
sentiments = df_cleaned[willingness_to_pay_column].value_counts()

# Define the categories you want to keep
categories_to_keep = ['Yes', 'No', 'Maybe']

# Sum the counts of all categories not in categories_to_keep
other_count = sentiments[~sentiments.index.isin(categories_to_keep)].sum()

# Remove categories not in categories_to_keep and add Others category
sentiments = sentiments[sentiments.index.isin(categories_to_keep)]
sentiments['Others'] = other_count

# Define colors for each category
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightgrey']

# Define label distances for each category
label_distances = [0.6, 0.6, 0.6, 0.5]  # Adjust the label distance for 'Others'

# Plotting the sentiments as a donut chart
plt.figure(figsize=(8, 6))
plt.pie(sentiments, labels=sentiments.index, colors=colors, autopct='%1.1f%%', startangle=90, pctdistance=0.85, wedgeprops=dict(width=0.4), labeldistance=1.1)
plt.title("Respondents' Sentiment towards Paying for Accommodation Search Apps")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Add a circle at the center to create the donut effect
centre_circle = plt.Circle((0, 0), 0.3, color='white', linewidth=0)
plt.gca().add_artist(centre_circle)

# Show the plot
plt.show()
```
